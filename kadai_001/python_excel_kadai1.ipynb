```python
import openpyxl
from datetime import datetime

#新しくワークブックを作成
workbook = openpyxl.Workbook()
ws = workbook.active

#会社情報を入力
ws['A2'] = "請求書"
ws['A4'] = "株式会社ABC"
ws['A5'] = "〒101-0022 東京都千代田区神田練塀町300"
ws['A6'] = "TEL:03-1234-5678 FAX:03-1234-5678"
ws['A7'] = "担当者名:鈴木一郎 様"
ws['E4'] = "No"
ws['E5'] = "日付"
ws['F4'] = "0001"

#現在の日付を取得
today = datetime.today()
ws['F5'] = f"{today.strftime('%Y/%m/%d')}"

#商品購入表を書き込む
salse_data = [
    ["商品名","数量","単価","金額"],
    ["商品A",2,10000,20000],
    ["商品B",1,15000,15000]
]

#合計金額を算出
total_amount = sum(item[3] for item in salse_data[1:])
#消費税額を算出
tax_amount = total_amount * 0.1

#購入情報をfor文でワークシートに追加する
for row in salse_data:
  ws.append(row)

#レイアウトの調整
ws.insert_rows(8,2)
ws.insert_cols(1,1)

ws['E13'] = total_amount

amount_cell = ws.max_row +2

#金額をワークシートに追加する
ws['B15'] = "合計"
ws['B16'] = "消費税"
ws['B17'] = "税込合計"

ws['E15'] =  total_amount
ws['E16'] =  tax_amount
ws['E17'] =  total_amount + tax_amount

#ファイルを保存
workbook.save(f'請求書_{today.strftime('%Y%m%d')}.xlsx')
```
