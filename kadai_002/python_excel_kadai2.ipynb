```python
import pandas as pd

#データフレームの作成
df = pd.DataFrame({
    '日付':['2023-05-17','2023-05-18','2023-05-19','2023-05-20','2023-05-21'],
    '社員名':['山田','佐藤','鈴木','田中','高橋'],
    '売上':[100,200,150,300,250],
    '部門':['メーカー','代理店','メーカー','商社','代理店']
})

#平均売上を算出し新しい列を作成
df['平均売上'] = df['売上'].mean()

#平均売上で分岐して業績ランクを設定する関数「performance」を定義
def performance(row):
    if row['売上'] >= row['平均売上'] + 50:
        return 'A'
    elif row['売上'] >= row['平均売上']:
        return 'B'
    else:
        return 'C'

# 「業績ランク」列を作成し、関数「performance」を適用して値を設定
df['業績ランク'] = df.apply(performance, axis=1)

# Excelファイルを作成
writer = pd.ExcelWriter('業績.xlsx')

# DataFrameオブジェクトをExcelファイルに読み込む
df.to_excel(writer,sheet_name='sheet1',index=False)

#Excelファイルを閉じる
writer.close()
```
