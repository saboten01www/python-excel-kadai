```python
import pandas as pd
import openpyxl
from openpyxl.styles import PatternFill,Border, Side

# Step3.データの連結と集計
# 処理対象のファイル名のリストを格納
file_list = ['2022_年間売上表.xlsx','2023_年間売上表.xlsx']
#結合後のデータを格納する空のデータフレームを作成
df = pd.DataFrame()

# ファイルを一つずつ処理する
for file_name in file_list:
  # Excelファイルを読み込む
  temp_df = pd.read_excel(file_name)
  #データフレームの列名を指定する
  temp_df.columns = ['売上年','商品','金額(千円)']
  # concat()でデータフレームを連結
  df = pd.concat([df, temp_df], ignore_index=True)

#「商品」「売上年」ごとにグループ化し、合計金額に集約
grouped = df.groupby(['商品','売上年']).agg({'金額(千円)': 'sum'})

# Step4. 売上集計表の作成とヘッダーの書式変更
# Exelファイルを作成する
writer = pd.ExcelWriter('売上集計表.xlsx')

# DataFrameオブジェクトをExcelファイルに書き込む
grouped.to_excel(writer,sheet_name='合計金額', index=True)

# Excelファイルを閉じる
writer.close()

# Excelファイルを読み込む
workbook = openpyxl.load_workbook('売上集計表.xlsx')

sheet = workbook.active

    # Side属性で線のスタイルを指定
thin_border = Side(style='thin')

# Border属性で各辺の線のスタイルを指定
border = Border(top=thin_border, bottom=thin_border, left=thin_border, right=thin_border)

# A1からC17の範囲に枠線を適用
for row_idx in range(1, 18): # Rows 1 to 17
    for col_idx in range(1, 4): # Columns A, B, C (1, 2, 3)
        cell = sheet.cell(row=row_idx, column=col_idx)
        cell.border = border

#フォントを設定する
for col_idx in range(1, 4): # Columns A, B, C (1, 2, 3)
    cell = sheet.cell(row=1, column=col_idx)
    cell.fill = PatternFill(patternType='solid', fgColor='FFF2F2F2')



# Excelファイルを保存する
workbook.save('売上集計表.xlsx')
```
